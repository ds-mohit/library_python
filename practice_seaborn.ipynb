```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

making different types of graph


```python

df= sns.load_dataset('tips')
df.head()
```

### 1. scatter plot

```python
sns.scatterplot(data=df,x="tip",y="total_bill")
plt.show()
```

```python
sns.scatterplot(data=df,x="tip",y="total_bill",hue="sex")
#hue: Splits data into different colors based on a categorical variable.
#  Each unique value in the column gets a different color.


```

```python
sns.scatterplot(data=df,x="tip",y="total_bill",hue="sex",style="smoker",palette="magma")
#style: Changes the marker style (shape) based on a categorical variable.
#palette: Sets the color scheme for the plot. 
# You can use built-in palettes (like "magma", "viridis", "deep", etc.) or
#  define your own list of colors.(like ["red", "blue", "green"])


```

```python
sns.scatterplot(data=df,x="tip",y="total_bill",hue="sex",style="smoker",palette="magma",size="size")
# size is a categorical variable which maps to the size of the marker
# it gives the size of the marker based on the value of the size variable
```

```python
sns.scatterplot(data=df,x="tip",y="total_bill",hue="sex",style="smoker",palette="magma",size="size",edgecolor="black")
#edgecolor is used to color the edges of the marker
```

```python
sns.scatterplot(
    data=df,
    x="tip",
    y="total_bill",
    hue="sex",
    style="smoker",
    size="size",
    palette=["red", "blue"],
    # alpha=0.7,
    # sizes=(40, 200),
    edgecolor="black",
    linewidth=0.5 # linewidth is the width of the line around the marker
    )
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="tip",
    y="total_bill",
    hue="sex",
    style="smoker",
    size="size",
    palette="cool",
    alpha=0.7,# is used to make the marker transparent
    sizes=(40, 200),# is used to define the range of the marker size
    edgecolor="black",
    linewidth=1
)
plt.show()
```

### 2. line plot

```python
sns.lineplot(
    data=df,x='tip',y='total_bill',hue='sex',palette=["red","blue"],#style="smoker",
)
```

```python
sns.lineplot(
    data=df,x='tip',y='total_bill',hue='sex',palette=["red","blue"],
    style="smoker",linewidth=0.5
    #in lineplot we can't  use size,edgecolor,markershape,markersize,etc
)
```

### 3. bar plot

```python
sns.barplot(data=df,x='sex',y='total_bill')
```

```python
sns.barplot(data=df,x='sex',y='total_bill',hue='smoker',palette=["red","blue"],alpha=0.9
            )
            #style="smoker" we can't use style in barplot
            # egdecolor,size,linewidth,marker,etc can't be used in barplot
```

### 4. histogram

```python
sns.histplot(data=df,x='total_bill',hue='sex',palette=["red","blue"],)
```

```python
sns.histplot(data=df,x='total_bill',hue='sex',palette=["red","blue"],
             bins=10#bins is used to show the no of times a data is divided
    #if bins are less it will show less details whreas more bins will show more details
             )
```

#### we can use kde in histplot

```python
sns.histplot(data=df,x='total_bill',hue='sex',kde=True,palette=["blue","yellow"],)
```

### 5. kde plot

```python
sns.kdeplot(data=df,x='total_bill',hue='sex',)
#kdeplot is used to show the density of the data
# it is the advanced version of histplot which is used to show the density of the data
```

### 5. box plot

```python
sns.boxplot(data=df,x='sex',y='total_bill',hue='sex',palette="magma",
            order=["Female","Male"] # order is used to show the order of the data
            )
# Box plots mainly support x, y, hue, palette, order, hue_order, orient, 
# and some statistical options.
```

### 6. violin plot

```python
sns.violinplot(data=df,x='sex',y='total_bill',hue='sex',)
#violinplot mainly supports x, y, hue, palette, order, hue_order, orient, split,
#  scale, inner, and some statistical options.
```

```python
sns.violinplot(data=df,x='sex',y='total_bill',hue='sex',orient="h")
```

```python
sns.violinplot(data=df, x='sex', y='total_bill', hue='smoker', split=True)
```

#### 7. count plot


```python
sns.countplot(data=df,x='day',hue='smoker',palette="magma")
```

#### 8. pair plot

```python
sns.pairplot(data=df)
```

#### 9. categorical plot

```python
sns.catplot(data=df,x='total_bill',y='day')
```

```python
sns.catplot(x="day", y="total_bill", data=df, kind="swarm")
#"strip" (default): strip plot (scatter of points)
# "swarm": swarm plot (non-overlapping scatter)
# "box": box plot
# "violin": violin plot
# "boxen": boxen plot (letter-value plot)
# "point": point plot (mean and confidence interval)
# "bar": bar plot (mean and confidence interval)
# "count": count plot (counts of observations)
```

#### 10. rel plot(in this wecan use only scatter and line plot)

```python
sns.relplot(data=df,x='tip',y='total_bill',hue='sex',palette="magma",size="size")
# if we want to make line plot we use kind = "line" for line plot
```

```python
sns.relplot(data=df,x='tip',y='total_bill',kind="line",
            hue='sex',palette="magma",size="size"
            )
```

```python
sns.relplot(x="total_bill", y="tip", data=df,kind="scatter", hue="smoker",style="size",col="day")
```

```python
sns.relplot(x="total_bill", y="tip", data=df,kind="scatter", hue="smoker",style="size",row="day")
```

```python
##outliers handle
q1 = df["total_bill"].quantile(0.25)
q3 = df["total_bill"].quantile(0.75)
iqr = q3-q1
lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr
print(lower_limit,upper_limit)
df[df["total_bill"]>upper_limit]

df = df[(df["total_bill"]>lower_limit)&(df["total_bill"]<upper_limit)]
sns.boxplot(df["total_bill"])
```
