### Pandas

```Pandas``` is a Python library used for working with data sets.

It has functions for analyzing, cleaning, exploring, and manipulating data.

The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

### Why Use Pandas?
```Pandas``` allows us to analyze big data and make conclusions based on statistical theories.

Pandas can clean messy data sets, and make them readable and relevant.

Relevant data is very important in data science.

````python
df.loc[0] #get the first row of the data frame using loc
df.loc[1:10, ['bike_name', 'age']] #get rows 1 to 10 and columns 'bike_name' and 'age'
````

````python
print(df['price'].min())#get the minimum value of the column 'sell_price'
print(df['price'].max())#get the maximum value of the column 'sell_price'
print(df['price'].mean())#get the mean value of the column 'sell_price' 
print(df['price'].median())#get the median value of the column 'sell_price'
print(df['price'].std())#get the standard deviation of the column 'sell_price'
print(df['price'].unique())#get the unique values of the column 'sell_price'
print(df['price'].nunique())#get the number of unique values of the column 'sell_price'
print(df['price'].describe())#get the descriptive statistics of the column 'sell_price'
print(df['city'].value_counts())#get the count of each unique value of the column 'sell_price'
print(df.columns)
````

````python
# we want data of bikes that are in delhi
delhi = df[df['city'] == 'Delhi']  # filter the data frame for bikes in Delhi
print(delhi)
````

````python
#if we want to print data in which bike should be from delhi and the owner is "Fourth Owner"
delhi['owner'].value_counts()
delhi[delhi['owner'] == 'Fourth Owner Or More']  # filter the data frame for bikes in Delhi and owned by "Fourth Owner"

````

````python
# bikes which are run 65k km less or equal
delhi[delhi['kms_driven']>=60000.0]
````

````python
# delhi ,age =25,model=yamaha
delhi['age'].value_counts()
delhi['brand'].value_counts()
delhi[(delhi['age'] <= 25) & (delhi['brand'] == 'Yamaha')]
````

````python
df.duplicated().sum()  # check for duplicate rows in the data frame
#to delete duplicate values
df.drop_duplicates()  # remove duplicate rows from the data frame
df.duplicated().sum() #it still showing duplicate values because we have not assigned the result to a variable or used inplace=True
df.drop_duplicates(inplace=True)  # remove duplicate rows from the data frame and assign the result to the same variable
df.duplicated().sum()  # check for duplicate rows in the data frame again

````

````python
# null values
df.isnull().sum()  # check for null values in the data frame
df.dropna()  # remove rows with null values from the data frame
df.dropna(inplace=True)  # remove rows with null values from the data frame and assign the result to the same variable
````

````python
#if we want to create a new file after performing all the operations
df.to_csv('Cleaned_Used_Bikes.csv')  # save the cleaned data frame to a new CSV file
````
